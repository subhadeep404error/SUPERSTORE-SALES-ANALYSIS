# Monthly Sales 

```python
import calendar 
```

```python
df.columns
```

```python
df['Month'] = df['order_date'].dt.month
df['MonthName'] = df['Month'].apply(lambda x: calendar.month_name[x])
```

```python
monthly_sales=df.groupby('MonthName')['sales'].sum()
monthly_sales
```

```python
monthly_sales_sorted=monthly_sales.sort_values(ascending=False).reset_index()
monthly_sales_sorted
```

```python
plt.figure(figsize=(15,5))
az=sns.barplot(x='sales',y='MonthName',data=monthly_sales_sorted,edgecolor='black',linewidth=.5)
for bars in az.containers:
    az.bar_label(bars)
plt.xlabel('SALES',fontsize=20)
plt.ylabel('MONTHNAME',fontsize=20)
plt.title('MONTHLY SALES',fontweight='bold')
plt.show()
```
